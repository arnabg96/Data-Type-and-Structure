What are data structures, and why are they important

Data structures are ways of organizing and storing data so that they can be accessed and modified
efficiently. They are important because they provide a means to manage large amounts of data efficiently,
such as large databases and internet indexing services.

 Explain the difference between mutable and immutable data types with examples

Mutable data types can be changed after creation (e.g., lists, dictionaries), while immutable types cannot
(e.g., tuples, strings). Example: list1 = [1, 2]; list1[0] = 5 (mutable). str1 = 'hello'; str1[0] = 'H' (raises error as
strings are immutable).

What are the main differences between lists and tuples in Python

Lists are mutable, tuples are immutable. Lists have more built-in methods than tuples. Tuples can be
used as dictionary keys, lists cannot.

 Describe how dictionaries store data

Dictionaries store data in key-value pairs using a hash table for efficient lookups.


Why might you use a set instead of a list in Python

Sets automatically eliminate duplicates and offer faster membership tests compared to lists.

Why might you use a set instead of a list in Python

How do tuples ensure data integrity in Python

Since tuples are immutable, their values cannot be changed after creation, ensuring data integrity

What is a hash table, and how does it relate to dictionaries in Python

A hash table stores data using a calculated hash of the key. Python dictionaries use hash tables for fast
key-based access

Can lists contain different data types in Python

Yes, Python lists can contain elements of different data types, like integers, strings, etc.


Explain why strings are immutable in Python

Strings are immutable for performance and security reasons. Once created, they cannot be altered.

What advantages do dictionaries offer over lists for certain tasks

Dictionaries allow fast access to values using keys, whereas lists require searching through indices.


Describe a scenario where using a tuple would be preferable over a list

Tuples are preferable when the data should not change, such as coordinates or RGB color values

How do sets handle duplicate values in Python

Sets automatically remove duplicate values upon creation

How does the 'in' keyword work differently for lists and dictionaries

'in' checks for value presence in lists, but for dictionaries it checks for key presence

Can you modify the elements of a tuple? Explain why or why not

No, tuples are immutable, so their elements cannot be modified after creation.

What is a nested dictionary, and give an example of its use case

A nested dictionary is a dictionary within another dictionary. Example: {'emp': {'name': 'John', 'age': 30}}


Describe the time complexity of accessing elements in a dictionary

The average time complexity is O(1) due to hashing

In what situations are lists preferred over dictionaries

Lists are preferred when order and duplicate elements are important or when iterating over values in
order

Why are dictionaries considered unordered, and how does that affect data retrieval

Until Python 3.7, dictionaries were unordered. Now, they preserve insertion order, but logically they
allow fast key-based access, not sequential

Explain the difference between a list and a dictionary in terms of data retrieval

Lists use index-based retrieval (O(1)), while dictionaries use key-based retrieval (O(1) on average).
